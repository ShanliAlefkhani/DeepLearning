```python
import torch
import cv2
import numpy as np

from matplotlib import pyplot as plt
```

```python
a = torch.tensor([1, 2, 3])
print(a)
print(a.device)
```

```python
a = torch.tensor([10., 2., 3.], requires_grad=True)
print(a.is_leaf)
b = torch.tensor([1., 2., 3.], requires_grad=True)
print(b.is_leaf)

c = a * b

c.retain_grad()

d = 2 * c

d.retain_grad()

print(d.shape, c.shape)

d.backward(torch.tensor([1, 1, 1]))

print(c.grad)

print(d.grad)
```

```python
img = cv2.imread('test1.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img)
plt.show()
```

```python
kernel = np.array([
    [-1, 0, +1],
    [-2, 0, +2],
    [-1, 0, +1],
])

kernel1 = np.array([
    [-1, -2, -1],
    [0, 0, 0],
    [+1, +2, +1],
])

edge_img = cv2.filter2D(img, -1, kernel)

edge_img1 = cv2.filter2D(img, -1, kernel1)

print(edge_img.shape)

plt.imshow(edge_img)
plt.show()
plt.imshow(edge_img1)
plt.show()
```
